```python
import pandas as pd
```

```python
df = pd.read_csv('currency_data\\euro-daily-hist_1999_2022.csv')
df
```

```python
cols=['Date','USD']
data = df[cols]
data
```

```python
index_list = data.loc[data['USD']=='-'].index.to_list()
index_list
```

```python
data = data.drop(index=index_list)
data 
```

```python
data['USD']=pd.to_numeric(data['USD'])
```

```python
data = data.sort_values(by='Date',ascending=True)
data
```

```python
data['increase'] = data['USD'].shift(-1)-data['USD']
data['perc_inc'] = 100*data['increase']/data['USD']
data
```

```python
import seaborn as sns
```

```python
sns.boxplot(x=data['perc_inc'])
```

```python
q25_increase = data['increase'].quantile(0.25)
q75_increase = data['increase'].quantile(0.75)
```

```python
def encode_increase(increase):
    if increase<0:
        return 0
    elif increase==0:
        return 1
    elif increase>0:
        return 2

    
```

```python
data['increase_classes']=data['increase'].apply(encode_increase)
data
```
