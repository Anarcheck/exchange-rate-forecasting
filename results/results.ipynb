```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
gpt_small_acc = np.load('gpt_small_accuracy.npy')
gpt_medium_acc = np.load('gpt_medium_accuracy.npy')
gpt_big_acc = np.load('gpt_big_accuracy.npy')

gpt_small_loss = np.load('gpt_small_loss.npy')
gpt_medium_loss = np.load('gpt_medium_loss.npy')
gpt_big_loss = np.load('gpt_big_loss.npy')
```

```python
gpt_small_acc = gpt_small_acc * 100
gpt_medium_acc = gpt_medium_acc * 100
gpt_big_acc = gpt_big_acc * 100
print(gpt_small_acc)
print(gpt_medium_acc)
print(gpt_big_acc)
```

```python
x1 = [i for i in range(1,gpt_small_acc.size+1)]
x2 = [i for i in range(1,gpt_medium_acc.size+1)]
x3 = [i for i in range(1,gpt_big_acc.size+1)]
```

```python
fig, axes = plt.subplots(3)
axes[0].plot(x1, gpt_small_acc)
axes[0].set_title('GPT Small Model')
axes[0].set_xlabel('Epochs')
axes[0].set_ylabel('Accuracy')

#  
axes[1].plot(x2, gpt_medium_acc)
axes[1].set_title('GPT Medium Model')
axes[1].set_xlabel('Epochs')
axes[1].set_ylabel('Accuracy')

#  
axes[2].plot(x3, gpt_big_acc)
axes[2].set_title('GPT Big Model')
axes[2].set_xlabel('Epochs')
axes[2].set_ylabel('Accuracy')

plt.tight_layout()
plt.show()
```

```python
plt.figure()
plt.plot(x3, gpt_big_acc, label='Accuracy', color='blue')
plt.legend()
plt.show()
```

```python
plt.figure()
plt.plot(x3, gpt_big_loss, label='Loss', color='blue')
plt.legend()
plt.show()
```

```python
print(np.round(gpt_big_acc.max(),3))
print(gpt_big_loss.mean())
```
