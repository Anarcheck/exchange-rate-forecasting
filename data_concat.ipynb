```python
import pandas as pd
import string
```

```python
data1 = pd.read_csv('data_1.csv', index_col=False)
data1 = data1.drop(columns=['Unnamed: 0'])
data1
```

```python
data2 = pd.read_csv('data_2.csv')
data2 = data2.drop(columns=['Unnamed: 0'])
data2
```

```python
data3 = pd.read_csv('data_3.csv')
data3 = data3.drop(columns=['Unnamed: 0'])
data3
```

```python
data4 = pd.read_csv('data_4.csv')
data4 = data4.drop(columns=['Unnamed: 0'])
data4
```

```python
data5 = pd.read_csv('data_5.csv')
data5 = data5.drop(columns=['Unnamed: 0'])
data5
```

```python
dataset = pd.concat([data1, data2, data3, data4, data5], ignore_index=True)
dataset
```

```python
na_counts = dataset.isna().sum()
na_counts
```

```python
dataset = dataset.dropna() 
```

```python
dataset = dataset.reset_index(drop=True)
dataset
```

```python
dataset = dataset.sort_values(by='Date', ascending=True)
dataset
```

```python
dataset['clean text'] = dataset['Title'].str.translate(str.maketrans('', '', string.punctuation)).str.lower()
dataset
```

```python
grouped_df = dataset.groupby('Date')['clean text'].apply(', '.join).reset_index()
grouped_df
```

```python
currency_data=pd.read_csv('currency_data.csv')
currency_data=currency_data.drop(columns=['Unnamed: 0'])
currency_data
```

```python
merged_data=pd.merge(grouped_df,currency_data,on='Date')
```

```python
merged_data
```

```python
merged_data.to_csv('merged_data.csv')
```
