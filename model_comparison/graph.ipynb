```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
epochs = [i for i in range(1,21)]
```

```python
gpt_acc = np.load('gpt2_accuracy.npy')
bert_acc = np.load('bert_accuracy.npy')
roberta_acc = np.load('roberta_accuracy.npy')

gpt_acc = gpt_acc * 100
bert_acc = bert_acc * 100
roberta_acc = roberta_acc * 100

```

```python
fig, axes = plt.subplots(3)
axes[0].plot(epochs, gpt_acc)
axes[0].set_title('GPT Model')
axes[0].set_xlabel('Epochs')
axes[0].set_ylabel('Accuracy')

# Второй подграфик
axes[1].plot(epochs, bert_acc)
axes[1].set_title('BERT Model')
axes[1].set_xlabel('Epochs')
axes[1].set_ylabel('Accuracy')

# Третий подграфик
axes[2].plot(epochs, roberta_acc)
axes[2].set_title('RoBERTa Model')
axes[2].set_xlabel('Epochs')
axes[2].set_ylabel('Accuracy')

plt.tight_layout()
plt.show()
```
